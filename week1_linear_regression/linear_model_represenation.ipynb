<h1>Univariant (One variable) Linear Regression</h1>

<h2>Problem Statement:</h2> In this lab, the units of size are 1000 sqft and the units of price are 1000s of dollars.

| Size (1000 sqft)     | Price (1000s of dollars) |
| -------------------| ------------------------ |
| 1.0               | 300                      |
| 2.0               | 500                      |

You would like to fit a linear regression model (shown above as the blue straight line) through these two points, so you can then predict price for other houses - say, a house with 1200 sqft.


<h2>Import libraries</h2>

```python
import numpy as np
import matplotlib.pyplot as plt
```

<h2>Create the input feature and the output target as numpy array</h2>

```python
x_train = np.array([1.0, 2.0])
y_train = np.array([300.0, 500.0])

print(f"x_train: {x_train}")
print(f"y_train: {y_train}")
```

<h2>Vitualize the data by plotting scatter plot using matplotlib</h2>

```python
plt.scatter(x_train, y_train, color='r')
plt.show()
```

<h2>Create a model to fit the data</h2>
<h3>
The model formula is give by the equation of a line:
$$f_{wb}(x) = wx^{(i)} + b \tag1$$
Where:</h3>
<h3>f_wb is the prediction, otherwise known as y_hat</h3>
<h3>x = input features, i ranges from 0 to m where m is the total number of features</h3>
<h3>w = weight</h3>
<h3>b = bias term</h3>


```python
def univariant_model(x,w,b):
    """
    Computes the prediction
    input: x ndarray
    parameters: w, b are weight and bias term respectively
    Ouput: f_wb or y_hat prediction    
    """
    # Get the total number of the input features ie m
    m = x.shape[0]
    # Initialize zeros to the prediction, f_wb = y_hat
    f_wb = np.zeros(m)
    # Loop m and compute the prediction
    for i in range(m):
        f_wb[i] = w * x[i] + b
    return f_wb    
```

<h2>Run the model</h2>

```python
# Assume weight = 200, b = 100, how this value are gotten will be explained later
y_hat = univariant_model(x_train, 200, 100)
print(f_wb)
```

<h2>Plot the original data together with its prediction</h2>

```python
plt.scatter(x_train, y_train, color='r')
plt.plot(x_train, y_hat)
plt.title("House Prediction")
plt.ylabel("Price (1000s of dollars)")
plt.xlabel("Size (1000 sqft)")
plt.show()
```

<h2>Now predict the price of a house with 1200 sqft</h2>

```python
#  Define a prediction function
def predict(x,w,b):
    return w * x + b    
```
